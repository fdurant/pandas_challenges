# Challenge 5

what director in your dataset has the highest gross per movie?

```python
moviesDFGroupedByDirector = moviesDF.groupby('Director')
dtgSeriesByDirector = moviesDFGroupedByDirector['DomesticTotalGross'].sum()
dtgSeriesByDirector.sort(axis=0, ascending=False)
dtgSeriesByDirector.head()
```

# Challenge 6

bin your dataset into years (if applicable) and make a bar graph with error bars of gross each year

```python
from datetime import datetime, date 
# Doing it by month instead of by year, for the sake of the exercise
moviesDFGroupedByMonth = moviesDF.groupby(lambda x: x.month)
dtgSeriesByMonth = moviesDFGroupedByMonth['DomesticTotalGross'].sum()
dtgStdByMonth = moviesDFGroupedByMonth['DomesticTotalGross'].std()
#dtgSeriesByMonth.values

plt.bar(np.array(dtgSeriesByMonth.index),dtgSeriesByMonth.values,yerr=dtgStdByMonth)
plt.title('Domestic Total Gross in 2013, per month')
plt.xlabel('Month')
plt.ylabel('Domestic Total Gross (USD)')
plt.show()
```
